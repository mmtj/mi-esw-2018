```python
import pandas
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
rcu_reads = pandas.read_csv('./rcu_reads.csv')
rcu_writes = pandas.read_csv('./rcu_writes.csv')

rwlock_reads = pandas.read_csv('./rwlock_reads.csv')
rwlock_writes = pandas.read_csv('./rwlock_writes.csv')
```

```python
rcu_reads_means = rcu_reads.mean()
rcu_writes_means = rcu_writes.mean()
rwlock_reads_means = rwlock_reads.mean()
rwlock_writes_means = rwlock_writes.mean()
```

```python
frames = {
    'rcu_reads': rcu_reads_means,
    'rcu_writes': rcu_writes_means,
    'rwlock_reads': rwlock_reads_means,
    'rwlock_writes': rwlock_writes_means
}
df = pandas.concat(frames, axis=1)
```

```python
xticks = range(0, 21)
yticks = range(0, 901, 100)

df.plot()
plt.title("RWLocks and RCU comparison")
plt.ylabel("Writes [K]")
plt.xlabel("Reader No.")
plt.yticks(yticks)
plt.xticks(xticks)
plt.savefig('rwlock_rcu.png')
plt.show()
```
